## Rolling Statistics continued: Exponential Weighted Average

An alternative way to calculate a rolling window is to take the exponentially weighted moving average or 'EWM'. This is similar to rolling std but using average. It takes the average amd puts more emphasis on the recent data rather than relying on older past data. This technique will help track trends, react to changes with new information, as well as balancing the past older data with the present data.

```python
# Use `ewm` to calculate the rolling window
rolling_ewa = df.ewm(span= 21).std()
rolling_ewa
```

# Sharpe Ratios
Now that we calculated some rolling statistics, its time to move on to another technique called Sharpe Ratio. Sharpe Ratio is process that will help manage the risk of a stock as well as the return to maximize the highest return rate with the lowest risk.


```python
annualization_factor = np.sqrt(252)
```

```python
df_mean = df.mean()
df_mean
```

```python
sharpe_r = (df_mean * annualization_factor)/ df_std

sharpe_r
```

```python
#Annualized Sharpe Ratios
sr = (df.mean()*252)/df.std() * np.sqrt(252)
sr.sort_values(inplace=True, ascending = False)
sr
```

```python
print(len(sr.index))
print(len(sr))
```

```python
# Visualize the sharpe ratios as a bar plot
plt.figure(figsize=(10, 6))
plt.bar(sr.index, sr)
plt.xlabel('Portfolio')
plt.ylabel('Sharpe Ratio')
plt.title('Annualized Sharpe Ratios')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
